```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df_retail = pd.read_excel('http://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx')
df_retail.head()
```

```python
#Stripping spaces in the description column
df_retail['Description'] = df_retail['Description'].str.strip()
#Dropping rows that does not contain invoice number
df_retail.dropna(axis = 0, subset=['InvoiceNo'], inplace = True)
df_retail['InvoiceNo'] = df_retail['InvoiceNo'].astype('str')
#Remove credit transaction
df_retail = df_retail[~df_retail['InvoiceNo'].str.contains('C')]
```

```python
df_retail.describe()
```

```python
df_france = df_retail[df_retail["Country"] == "France"]
df_france.describe()
```

```python
basket_france = (df_france.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0)
                 .set_index('InvoiceNo'))

basket_france.tail()
```

```python
def sum_to_boolean(x):
    if x<=0:
        return 0
    else:
        return 1

basket_france_sg = basket_france.applymap(sum_to_boolean)
```

```python
basket_france_sg.drop('POSTAGE', inplace=True, axis=1)
basket_france_sg.describe()
```

```python
frequent_itemsets = apriori(basket_france_sg, min_support = 0.05, use_colnames = True)

frequent_itemsets.sort_values('support', ascending = False).head()
```

```python
frequent_itemsets.sort_values('support', ascending = False).tail()
```

```python
confidence = association_rules(frequent_itemsets, metric = "confidence", min_threshold = 0.7)
confidence.sort_values('confidence',ascending = False)
```

```python
lift = association_rules(frequent_itemsets, metric = "lift", min_threshold = 1)
lift.sort_values('lift',ascending = False).head()
```

#### 'RABBIT NIGHT LIGHT' is the itemset with the largest support. 
#### The rule with the highest confidence is {'SET/20 RED RETROSPOT PAPER NAPKINS','SET/6 RED SPOTTY PAPER PLATES'} -> {'SET/6 RED SPOTTY PAPER CUPS'}.
#### The rule with the highest lift is {'PACK OF 6 SKULL PAPER CUPS'} -> {'PACK OF 6 SKULL PAPER PLATES'}
#### The rule with the highest confidence and the rule with the highest lift are different. This is because the lift takes into account the rule's joint support as well as the antecedent's and consequents' support.  Confidence only takes into account joint support of the rule and the antecedent's support.
