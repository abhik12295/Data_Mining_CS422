```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,precision_recall_fscore_support, confusion_matrix,accuracy_score
```

```python
data_iris = load_iris()
target_names=["setosa","versicolor","virginica"]
df_iris = pd.DataFrame(data_iris.data, columns = data_iris.feature_names)
df_iris['Species'] = data_iris.target
```

```python
df_iris.head()
```

```python
#train_test_split is used to randomly split the data into training and testing sets
X = df_iris[data_iris.feature_names]
Y = df_iris['Species']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

### Training model on the data with relationship between X(Attributes of Iris) and Y(Species of Iris)

```python
classifier = DecisionTreeClassifier(max_depth = 1, random_state = 0,min_samples_split=5,min_samples_leaf=2)
classifier.fit(X_train,Y_train)
Y_pred = classifier.predict(X_test)

print(confusion_matrix(Y_test,Y_pred))


#Classification report
print(classification_report(Y_test, Y_pred,digits=3))

#Macro, Micro, Weighted
print("\t\tPrecision "+"\t\t"+"Recall"+"\t\t"+" F1 Score"+"\tSupport")
print("Macro    :",precision_recall_fscore_support(Y_test, Y_pred, average='macro'))
print("Micro    :",precision_recall_fscore_support(Y_test, Y_pred, average='micro'))
print("Weighted :",precision_recall_fscore_support(Y_test, Y_pred, average='weighted'))



```

### Micro macro and weighted difference:
### 1. Micro: says the function to compute F1 by considering total true positives, false negatives, and false positives
### 2. Macro: the function to compute f1 score and return the average withouth considering the proportion for each label in dataset
### 3. weighted: says the function to compute f1 for each label and returns the average considering the proportion for each label in dataset
